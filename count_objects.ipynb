```python
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('1.jpg')

# Detect common objects
box, label, c_score = cv.detect_common_objects(image)


# Draw bounding boxes and labels (font size/color cannot be changed)
output = draw_bbox(image, box, label, c_score)

output = cv2.resize(output, (1080, 720))

plt.figure(figsize=(16, 9)) 

# Display the output image
plt.imshow(output)
plt.show()

# Count zebras
print('Similar objects available in photo : ' , label)

get_number = label.count('chair')
print('Number of chair:', get_number)
```
